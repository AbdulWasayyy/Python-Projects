## üß© How It Works:
The app starts a 5-second countdown after every keystroke.

If the timer hits 0 with no typing detected ‚Üí the entire text area clears.

A messagebox warns the user they stopped writing.

Typing again restarts the timer automatically.

## üí° Customization:
| Setting            | Description                     | Default        |
| ------------------ | ------------------------------- | -------------- |
| Inactivity timeout | How long before text is deleted | `5 seconds`    |
| Font               | Typing font style               | `Helvetica 14` |
| Colors             | Background/text colors          | Dark mode      |


````python
import tkinter as tk
from tkinter import messagebox
import threading

class DangerousWritingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("The Most Dangerous Writing App")
        self.root.geometry("800x600")
        self.root.configure(bg="#222")

        self.text = tk.Text(
            root, wrap="word", font=("Helvetica", 14),
            bg="#1e1e1e", fg="white", insertbackground="white", undo=True
        )
        self.text.pack(expand=True, fill="both", padx=20, pady=20)

        self.label = tk.Label(
            root, text="Keep typing... stop for 5 seconds and you lose everything!",
            bg="#222", fg="orange", font=("Arial", 12)
        )
        self.label.pack(pady=5)

        self.last_keypress_time = None
        self.timer_thread = None
        self.is_running = True

        self.text.bind("<Key>", self.on_keypress)
        self.start_timer()

    def on_keypress(self, event):
        self.last_keypress_time = self.root.after_cancel(self.last_keypress_time) if self.last_keypress_time else None
        self.reset_timer()

    def reset_timer(self):
        self.last_keypress_time = self.root.after(5000, self.clear_text)

    def clear_text(self):
        self.text.delete("1.0", tk.END)
        messagebox.showwarning("Oops!", "You stopped typing! All your work has been deleted.")
        self.reset_timer()

    def start_timer(self):
        self.reset_timer()


if __name__ == "__main__":
    root = tk.Tk()
    app = DangerousWritingApp(root)
    root.mainloop()
````
